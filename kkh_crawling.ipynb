### 기본: 사람인 채용정보 페이지에서, 추천 지역 아래의 회사명-제목을 40개 뽑아 Data Frame으로 생성하기

```python
# User Agent
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}
```

```python
# 사람인 채용정보 페이지
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

res = requests.get(url, headers = headers)

res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')
```

```python
a1 = soup.findAll('a', attrs = {'class' : 'str_tit'})
a1
```

```python
# 기업명과 제목을 추출할 반복문 생성
list = []
for a in a1:
    title1 = a.get_text().strip()
    print('기업명: {0}'.format(title1))
    list.append(title1)
```

```python
# 출력해 놓고 보니 기업명과 제목이 같이 추출되었습니다
# 기업명-제목이 하나씩 번갈아 등장하는 규칙이 있어서,
# 홀수번째와 짝수번째를 분리하면 기업명과 제목도 분리할 수 있을 것 같았습니다

print(list[::2])
print(list[1::2])
```

```python
#크롬드라이버 경로 설정
path = 'C:\\Users\\enigma\\Documents\\chromedriver'
from selenium.webdriver.chrome.service import Service

path = Service('C:\\Users\\enigma\\Documents\\chromedriver')
#드라이버 객체 생성
browser = webdriver.Chrome(service=path)
```

```python
# 커리어연세로 접속
browser.get('https://career.yonsei.ac.kr/')
```

```python
# 로그인 버튼 클릭
browser.find_element(By.XPATH, '/html/body/div/div/div[2]/div/ul/li[1]/a').click()
```

```python
# 채용정보 페이지로 이동
browser.find_element(By.XPATH, '//*[@id="gnb"]/ul/li[2]/a').click()
```

```python
# 회사명, 제목을 추출
# 1) 제목 뽑아내기
browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[1]/td[2]/a').text

# 첫번째 페이지의 첫번째 제목이 출력된다
```

```python
yonsei_title = []

for i in range(1, 21):
    temp = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[{}]/td[2]/a'.format(i)).text
    yonsei_title.append(temp)
```

```python
# 기업명 뽑아내기
browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[1]/td[1]').text

# 첫번째 기업명이 출력된다
```

```python
yonsei_firm = []
for i in range(1, 21):
    temp = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[{}]/td[1]'.format(i)).text
    yonsei_firm.append(temp)
    
yonsei_firm
```

```python
# 이로써 상위 20개 (1페이지)의 기업명/제목을 모두 얻었다
# 과제에서는 40개의 기업명/제목을 원하므로 2페이지를 탐색한다

browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/ul[3]/li[4]/a').click()
```

```python
# 2페이지에서도 1페이지에서와 동일한 작업을 반복한다
# 그전에 XPath를 한번만 체크해 보도록 한다

t1 = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[1]/td[2]/a').text
t1

f1 = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[1]/td[1]').text
print(f1, ":", t1)

# 기업명/제목이 잘 추출된다
```

```python
for i in range(1, 21):
    temp = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[{}]/td[1]'.format(i)).text
    yonsei_firm.append(temp)
    
for i in range(1, 21):
    temp = browser.find_element(By.XPATH, '//*[@id="ptfolCareerProgramSearch"]/div[4]/table/tbody/tr[{}]/td[2]/a'.format(i)).text
    yonsei_title.append(temp)
```
